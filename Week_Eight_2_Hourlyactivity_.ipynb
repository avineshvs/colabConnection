```python
import pandas as pd                       # Pandas for data manipulation and analysis
import numpy as np                        # NumPy for numerical operations
```

```python
pd.set_option('display.max_columns', None)
Cal = pd.read_csv("/content/drive/MyDrive/Data Analytics/Resources/Week Eight/data/hourlyCalories_merged.csv",index_col=0)
Int = pd.read_csv("/content/drive/MyDrive/Data Analytics/Resources/Week Eight/data/hourlyIntensities_merged.csv",index_col=0)
Steps = pd.read_csv("/content/drive/MyDrive/Data Analytics/Resources/Week Eight/data/hourlySteps_merged.csv",index_col=0)
```

```python
Cal.head(10)
```

```python
Int.head(10)
```

```python
Steps.head(10)
```

```python
print(Cal.shape)
print(Int.shape)
print(Steps.shape)
```

```python
print(Cal.isna().sum().sum())
print(Int.isna().sum().sum())
print(Steps.isna().sum().sum())
```

```python
HourlyActivity  = pd.merge(Cal,Int,how='inner',on=['Id','ActivityHour'])
HourlyActivity.head()
```

```python
HourlyActivity.shape
```

```python
HourlyActivity  = pd.merge(HourlyActivity,Steps,how='outer',on=['Id','ActivityHour'])
HourlyActivity.head()
```

```python
HourlyActivity.shape
```

```python
# Convert the 'date' column to datetime format
HourlyActivity['ActivityHour'] = pd.to_datetime(HourlyActivity['ActivityHour'], format='%m/%d/%Y %H:%M:%S %p')

```

```python
HourlyActivity.head()
```

```python
# Extract date and time into separate columns
HourlyActivity['ActivityDate'] = HourlyActivity['ActivityHour'].dt.date
HourlyActivity['ActivityTime'] = HourlyActivity['ActivityHour'].dt.time
HourlyActivity.head()
```

```python
HeartSave = '/content/drive/MyDrive/Data Analytics/Resources/Week Eight/ProcessedData/HourlyActivity.csv'
HourlyActivity.to_csv(HeartSave)
```
